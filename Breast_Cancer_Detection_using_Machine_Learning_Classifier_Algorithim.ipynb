```python
#Split  DataFrame in Train and Test
#input variable
X= cancer_df.drop(['target'],axis =1)
X.head()
```

```python
#output variable
y = cancer_df['target']
y.head(6)
```

```python
#split dataset into train and test
from sklearn.model_selection import  train_test_split
X_train,X_test,y_train,y_test  = train_test_split(X,y, test_size =0.2, random_state = 5)
```

```python
X_train
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc =  sc.transform(X_test)
```

**Machine Learning Model**

```python
from sklearn.metrics import confusion_matrix,classification_report ,accuracy_score
```

**Support Vector Classifier**

```python
from sklearn.svm import SVC
svc_classifier = SVC()
svc_classifier.fit(X_train_sc,y_train)
y_pred_scv =  svc_classifier.predict(X_test_sc)
accuracy_score(y_test,y_pred_scv)
```

**XGBoost Classifier**

```python
from xgboost import XGBClassifier
xgb_classifier = XGBClassifier()
model = xgb_classifier.fit(X_train_sc,y_train)
y_pred_xgb_sc = xgb_classifier.predict(X_test_sc)
accuracy_score(y_test,y_pred_xgb_sc)
```

```python
import pickle
```

```python
pickle.dump(model,open('breast_cancer_detector.pickle','wb'))
```
